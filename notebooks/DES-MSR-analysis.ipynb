```r
library(tidyverse)
library(magrittr)
library(zeallot)
source('../R/des-classic.R')
source("../R/des-msr.R")
source("../R/utils.R")
options(repr.plot.width=20, repr.plot.height=13)
```

```r
res_cls = run_experiment(optimalg = des_classic, eval = crossprod, dims = c(2, 3, 5, 10, 20, 30), x0 = 90)
```

```r
res_msr = run_experiment(optimalg = des_msr, eval = crossprod, dims = c(2, 3, 5, 10, 20, 30), x0 = 90, control = list(d_param = 10, c_param = 0.01))
```

```r
res = merge_results(list(res_cls, res_msr))
```

```r
res %>% vg_plot(yval = "best_pop", color_column = "label", scale_y_log = 1) + facet_wrap( ~ Dim)
```

```r
res_msr %>% vg_plot(yval = "sigma", color_column = "label", scale_y_log = 1) + facet_wrap( ~ Dim )
```
